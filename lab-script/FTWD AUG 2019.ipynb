```python

def CloneWeb(Dictionary=Students, Labs=Labs):
    
    for name, username in Dictionary.items(): 
        os.mkdir(name)
        path = os.getcwd() +'/'+ name 
        
        count = 0
        for lab in Labs: 
#             print("labs are being copied")
            try:
                count +=1 
                fork = lab.replace('ironhack-labs',username)
                lname = re.findall('lab-.*',lab)
                print(lname)
                print(fork)
                print(path)
                print(str(count))
#               Repo.clone_from(fork,path+'/'+str(count)+': '+lname)
                Repo.clone_from(fork,path+"/"+str(count) + ") " +  lname[0])
                print(fork,path+'/' + lname, 'Successfully Cloned')
            except: 
                print(fork,path+"/"+str(count) + ") " +  lname[0] +  ' does not exist')
                pass

```
