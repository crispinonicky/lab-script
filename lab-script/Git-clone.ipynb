```python
def Clone(Dictionary, subfolder):
    
    if os.path.isdir(subfolder) != True:  #check if subfolder exists and creates if not
        os.mkdir(subfolder)
    else: 
        pass
        
    for key, value in Dictionary.items():
        os.mkdir(subfolder+'/'+key)                               #make directory 
        path = os.getcwd()+'/'+subfolder+'/'+ key                 #Path of directory
        Repo.clone_from(value, path)                              #git clone
        print(key, 'Repo Successfully Cloned')                    #confirmation

        
def Pull(): 
    
    FT_S = [i for i in os.listdir('/Users/tompelletier/Ironhack/TA/FT_labs')]
    PT_S = [i for i in os.listdir('/Users/tompelletier/Ironhack/TA/PT_labs')]
    
    for i in FT_S: 
        if not i.startswith('.'):
            path = '/Users/tompelletier/Ironhack/TA/FT_labs'+'/'+i
            
            repo = git.Repo(path)
            repo.git.reset('--hard')
            
            g = git.cmd.Git(path)
            g.pull()
            print(i, 'Repo Pulled')
    print('\nALL FT repos Pulled\n')    
    for i in PT_S: 
        if not i.startswith('.'):
            path = '/Users/tompelletier/Ironhack/TA/PT_labs' +'/'+i
            
            repo = git.Repo(path)
            repo.git.reset('--hard')
            
            g = git.cmd.Git(path)
            g.pull()
            print(i, 'Repo Pulled')
    print('\nALL PT repos Pulled')  
```
